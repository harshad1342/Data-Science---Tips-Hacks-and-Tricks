## Pandas Boolean Indexing

```python
df = pd.read_csv("csv/loan_train.csv", index_col = 'Loan_ID')
df.head(5)
```

```python
# we want a list of all females who are not graduate and got a loan. 

df.loc[(df['Gender'] == 'Female') & (df['Education'] == 'Not Graduate') & (df['Loan_Status'] == 'Y'),
         ["Gender","Education","Loan_Status"]].head()
```

## Spliting string 

```python
# create a dataframe
df = pd.DataFrame({'name':['Elise Mccann', 'Aiden Berger', 'Elle Kelley']})
df
```

```python
# extract first name and last name
df['first_name'] = df['name'].str.split(' ', expand = True)[0]
df['last_name'] = df['name'].str.split(' ', expand = True)[1]

df
```

## Removing Emojis from Text

```python
text= "Hi 😂! Have a nice weekend 💕👭"
preprocessed_text=text.encode('ascii', 'ignore').decode('ascii')

print("Raw tweet:",text)  #with emoji
print("Preprocessed tweet:",preprocessed_text) # no emoji
```

## Time Series Data

```python
import datetime as dt
```

```python
df = pd.read_csv("csv/AirQualityUCI.csv")
df.head()
```

```python
df.dtypes
```

```python
df['date_time'] = df['Date'] + ' ' + df['Time']
df['date_time'] = pd.to_datetime(df['date_time'],format='%d/%m/%Y %H.%M.%S')
```

```python
new_df = pd.DataFrame({
    "date_time":df['date_time'],
    "year":df['date_time'].dt.year,
    "month":df['date_time'].dt.month,
    "date":df['date_time'].dt.date,
    "hour":df['date_time'].dt.hour,
    "CO(GT)":df['CO(GT)'],
    "PT08.S1(CO)":df['PT08.S1(CO)'],
    "NMHC(GT)":df['NMHC(GT)']
})
new_df.head()
```

```python
from tokenizers import (ByteLevelBPETokenizer,
                            BPETokenizer,
                            SentencePieceBPETokenizer,
                            BertWordPieceTokenizer)
#instantiate tokenizer
tokenizer = ByteLevelBPETokenizer()

#train with our own data
tokenizer.train(["sample.txt"], vocab_size=20000)

#Lets tokenize the sentence
output = tokenizer.encode("Hello! How are you?")

print("Unique id:",output.ids)
print("Tokens:",output.tokens)
print("Merges:",output.offsets)
```
